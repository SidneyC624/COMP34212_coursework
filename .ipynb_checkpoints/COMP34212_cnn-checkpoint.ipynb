```python
import itertools
from tensorflow.keras.datasets import cifar100
from tensorflow.keras import utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, BatchNormalization
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD, Adam, RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
N_EPOCH = 20
BATCH_SIZE = 32
IMG_ROWS = 32
IMG_COLS = 32
IMG_CHANNELS = 3
N_CLASSES = 100
VALIDATION_SPLIT = 0.2
VERBOSE = 1
REPEATS = 3
```

Loading and processing CIFAR-10 data

```python
#load dataset
(input_X_train, output_y_train), (input_X_test, output_y_test) = cifar100.load_data()
print('input_X_train shape:', input_X_train.shape)
print(input_X_train.shape[0], 'train samples')
print(input_X_test.shape[0], 'test samples')

# convert to categorical
output_Y_train = utils.to_categorical(output_y_train, N_CLASSES)
output_Y_test = utils.to_categorical(output_y_test, N_CLASSES)

# float and normalization
input_X_train = input_X_train.astype('float32')
input_X_test = input_X_test.astype('float32')
input_X_train /= 255
input_X_test /= 255
```

```python
# Hyperparameter configurations
learning_rates = [0.01, 0.0015, 0.0001]
filters_first_convs = [32, 96, 128]
dropout_rates = [0.3, 0.4, 0.7]

# Experimentation loop
results = {}
```

Defining CNN model

```python
# Fixed base values (choose any reasonable default)
base_lr = 0.001
base_filters = 64
base_dropout = 0.5

# Generate 9 configurations by varying only one parameter at a time
configurations = []

# Vary learning rate, keep others constant
for lr in learning_rates:
    configurations.append((lr, base_filters, base_dropout))

# Vary filter size, keep others constant
for filters in filters_first_convs:
    configurations.append((base_lr, filters, base_dropout))

# Vary dropout rate, keep others constant
for dropout in dropout_rates:
    configurations.append((base_lr, base_filters, dropout))

# Print configurations to check
print("Generated Configurations:")
for config in configurations:
    print(config)

for lr, filters, dropout in configurations:
    config_name = f"LR_{lr}_Filters_{filters}_Dropout_{dropout}"
    print(f"\nTraining configuration: {config_name}")

    accuracies = []

    for repeat in range(REPEATS):
        print(f"Run {repeat + 1} for {config_name}")

        # Build model
        model = Sequential()

        # 1st Convolutional Block
        model.add(Conv2D(filters, kernel_size=3, padding='same', input_shape=(IMG_ROWS, IMG_COLS, IMG_CHANNELS)))
        model.add(BatchNormalization())
        model.add(Activation('relu'))
        model.add(Conv2D(filters, kernel_size=3, padding='same'))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Dropout(dropout))

        # 2nd Convolutional Block
        model.add(Conv2D(128, kernel_size=3, padding='same'))
        model.add(BatchNormalization())
        model.add(Activation('relu'))
        model.add(Conv2D(128, kernel_size=3, padding='same'))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Dropout(dropout))

        # 3rd Convolutional Block
        model.add(Conv2D(256, kernel_size=3, padding='same'))
        model.add(BatchNormalization())
        model.add(Activation('relu'))
        model.add(Conv2D(256, kernel_size=3, padding='same'))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Dropout(dropout))

        # Flatten and Fully Connected Layers
        model.add(Flatten())
        model.add(Dense(1024))
        model.add(Activation('relu'))
        model.add(Dropout(dropout))
        model.add(Dense(N_CLASSES))
        model.add(Activation('softmax'))

        # Compile model
        optimizer = Adam(learning_rate=lr)
        model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

        # Train model
        history = model.fit(input_X_train, output_Y_train, batch_size=BATCH_SIZE,
                            epochs=N_EPOCH,
                            validation_split=VALIDATION_SPLIT,
                            verbose=VERBOSE)

        # Evaluate model
        score = model.evaluate(input_X_test, output_Y_test, batch_size=BATCH_SIZE, verbose=VERBOSE)
        accuracies.append(score[1])  # Store accuracy

    # Store results for this configuration
    results[config_name] = {
        "accuracies": accuracies,
        "mean_accuracy": np.mean(accuracies),
        "std_accuracy": np.std(accuracies)
    }

# Print summary of results
print("\nFinal Results:")
for config, metrics in results.items():
    print(f"\n{config}:")
    print(f"Mean Accuracy = {metrics['mean_accuracy']:.4f}, Std Dev = {metrics['std_accuracy']:.4f}")
    print(f"  5 Repetition Accuracies: {', '.join([f'{acc:.4f}' for acc in metrics['accuracies']])}")

# Plot results
configs = list(results.keys())
means = [results[c]["mean_accuracy"] for c in configs]
stds = [results[c]["std_accuracy"] for c in configs]

plt.figure(figsize=(12, 6))
plt.barh(configs, means, xerr=stds, capsize=5, color='skyblue', alpha=0.7)
plt.xlabel("Mean Test Accuracy")
plt.ylabel("Hyperparameter Configurations")
plt.title("Hyperparameter Tuning Results")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
import pandas as pd

results = {
    "LR_0.01_Filters_64_Dropout_0.5": [0.0100, 0.0100, 0.0100],
    "LR_0.0015_Filters_64_Dropout_0.5": [0.3523, 0.3509, 0.0100],
    "LR_0.0001_Filters_64_Dropout_0.5": [0.4899, 0.4963, 0.4890],
    "LR_0.001_Filters_32_Dropout_0.5": [0.4022, 0.4272, 0.3968],
    "LR_0.001_Filters_96_Dropout_0.5": [0.4424, 0.4162, 0.4254],
    "LR_0.001_Filters_128_Dropout_0.5": [0.4443, 0.4419, 0.4484],
    "LR_0.001_Filters_64_Dropout_0.3": [0.4664, 0.4699, 0.4658],
    "LR_0.001_Filters_64_Dropout_0.4": [0.4662, 0.4742, 0.4546],
    "LR_0.001_Filters_64_Dropout_0.7": [0.0100, 0.2227, 0.0100]
}

learning_rates = ["LR_0.01", "LR_0.0015", "LR_0.0001"]
filters = ["Filters_32", "Filters_96", "Filters_128"]
dropouts = ["Dropout_0.3", "Dropout_0.4", "Dropout_0.7"]

learning_rate_results = []
filter_results = []
dropout_results = []

for key, values in results.items():
    mean_acc = round(np.mean(values), 4)
    row = values + [mean_acc]
    
    if any(lr in key for lr in learning_rates):
        learning_rate_results.append(row)
    elif any(f"Filters_{f}" in key for f in [32, 96, 128]):
        filter_results.append(row)
    elif any(f"Dropout_{d}" in key for d in [0.3, 0.4, 0.7]):
        dropout_results.append(row)

# Save tables as CSV files
def save_table(data, filename, index_labels):
    df = pd.DataFrame(data, columns=["Run 1", "Run 2", "Run 3", "Mean Accuracy"], index=index_labels)
    df.to_csv(filename)

save_table(learning_rate_results, "learning_rate_results.csv", learning_rates)
save_table(filter_results, "filter_results.csv", filters)
save_table(dropout_results, "dropout_results.csv", dropouts)

# Summary table
# round to 4 decimal places
summary_data = [
    [round(row[-1], 4) for row in learning_rate_results],
    [round(row[-1], 4) for row in filter_results],
    [round(row[-1], 4) for row in dropout_results]
]
summary_df = pd.DataFrame(summary_data, columns=["Valuation 1", "Valuation 2", "Valuation 3"], 
                           index=["Learning Rate (0.01, 0.0015, 0.0001)", "Filters (32, 96, 128)", "Dropout (0.3, 0.4, 0.7)"])
summary_df.to_csv("summary_results.csv")

# Error bar plot
labels = list(results.keys())
means = [np.mean(acc) for acc in results.values()]
stdevs = [np.std(acc) for acc in results.values()]

plt.figure(figsize=(12, 6))
plt.errorbar(labels, means, yerr=stdevs, fmt='o', capsize=5, markersize=8, label="Accuracy")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Hyperparameter Configuration")
plt.ylabel("Accuracy")
plt.title("Hyperparameter Tuning Results with Error Bars")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("error_bar_plot.png")
plt.show()
```

```python
# Function to generate and save table images
def save_table_as_image(csv_file, image_file, title):
    df = pd.read_csv(csv_file, index_col=0)
    
    fig, ax = plt.subplots(figsize=(10, len(df) * 0.5 + 2))
    ax.axis("tight")
    ax.axis("off")
    table = ax.table(cellText=df.values, colLabels=df.columns, rowLabels=df.index, 
                      cellLoc="center", loc="center")

    table.auto_set_font_size(False)
    table.set_fontsize(10)  # Set font size manually
    table.scale(1, 1.5)  # Increase row height (second parameter)

    plt.title(title, fontsize=14, fontweight="bold")
    plt.savefig(image_file, dpi=300, bbox_inches="tight")
    plt.close()

# Generate table images
save_table_as_image("learning_rate_results.csv", "learning_rate_results.png", "Learning Rate Results")
save_table_as_image("filter_results.csv", "filter_results.png", "Filter Results")
save_table_as_image("dropout_results.csv", "dropout_results.png", "Dropout Results")
save_table_as_image("summary_results.csv", "summary_results.png", "Summary Results")

print("Table images saved successfully.")
```
